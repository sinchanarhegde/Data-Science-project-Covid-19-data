## Filtering of data:

```python
a=df2.iloc[:,2:15]
```

```python
a
```

```python
b=df2.iloc[:,-1]
```

```python
b
```

### Model Training:

```python
from sklearn.model_selection import train_test_split
```

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.25,random_state=(0))
```

### Logistic Regression:

```python
from sklearn.linear_model import LogisticRegression
```

```python
c=LogisticRegression(max_iter=10000)
c.fit(a_train,b_train)
```

```python
c_predict=c.predict(a_test)
```

### Confusion Matrix:

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
acs=accuracy_score(b_test,c_predict)
cm=confusion_matrix(b_test,c_predict)
```
